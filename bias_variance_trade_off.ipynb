# Bias-Variance trade off

**Bias**: The inability for a machine learning method (like linear regression) to capture the true relationship is called bias.

**Variance**: In machine learning, the difference in between data sets is called variance.
>The underfitting models have high bias and low variance whereas the overfitting models possess high variance and low bias.

**Bias-Variance tradeoff**: the balancing act that is required between models that are too simple and those that are too complex


---


imagine we have a dataset where data points scattered around an imaginary curve (e.g an logarithmic curve),  
if our model simply just draw a straight line like the ones in linear regression, our model would have **low variance** but **high bias** as it cannot capture the curve in the relationship

in example below imagine such relation is true for height and weight of mice, a straight line would have a huge MSE

```python
from matplotlib import pyplot as plt
import numpy as np


height = range(1, 11)
actual_weight = np.array([1, 10, 20, 39, 47, 18, 53, 52, 50, 51])
predicted_weight = np.array([7 * i for i in range(10)])

plt.scatter(height, actual_weight)
plt.plot(height, predicted_weight, 'red')

print("mean square error is: ", np.mean((actual_weight- predicted_weight)**2))

plt.show()
```

how ever if we make our model **too complicated** it can cover all data point (even that i[5] which an outlier),
imagine a squiggly line which will pass through all these points like this:

```python
from matplotlib import pyplot as plt
import numpy as np


height = range(1, 11)
actual_weight = np.array([1, 10, 20, 39, 47, 18, 53, 52, 50, 51])
squiggly_line = actual_weight

plt.scatter(height, actual_weight)
plt.plot(height, squiggly_line, 'red')

print("mean square error is: ", np.mean((actual_weight- squiggly_line)**2))

plt.show()
```

in the second example our model is too complicated but it has no error in training set, however the previous model is more likely to perform better in test set,
